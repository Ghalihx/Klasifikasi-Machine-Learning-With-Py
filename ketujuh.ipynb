#Linear Regression

```python
x = data [['Hours_Studied','Attendance','Sleep_Hours','Previous_Scores']] #Features
y = data[['Exam_Score']] #Target
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

#Split Data
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)

#Training
model = LinearRegression()
model.fit(x_train, y_train)

#Evaluasi
y_pred = model.predict(x_test)

mse = mean_squared_error(y_pred,y_test)
print(f'Hasil Mean Squared Error = {mse:.3f}')
```

```python
model.coef_
```

Feature Importance

```python
import numpy as np
feature_importance = pd.DataFrame({
    'Feature': x.columns,
    'Importance': model.coef_[0]
})
feature_importance = feature_importance.sort_values(by='Importance',ascending = False)
print(feature_importance)
```

Berdasarkan Hasil Feature Importance dari model Linear Regression, dapat dilihat bahwa Hours_Studied (Jam belajar Perminggu) memiliki nilai kepentingan (importance) yang paling tinggi, sehingga menunjukan  pengaruh paling besar terhadap Exam Score (Nilai Ujian). Attendance(kehadiran) juga memberikan pengaruh terhadapat Exam Score, namun tidak tinggi dibandingkan Hours_Studied, Begitu juga dengan Previous_Scores.

Sedangkan Sleep Hours dikarenakan memberikan hasil berupa minus pada nilai importance sehingga tidak memberikan pengaruh pada prediksi model atau terhadap Exam Scores.

#Klasifikasi / Classification

```python
#Label Encoding
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
data2['Parental_Involvement_encoded'] = label_encoder.fit_transform(data['Parental_Involvement'])
data2
```

```python
label_mapping = dict(zip(label_encoder.classes_, label_encoder.transform(label_encoder.classes_)))
print(label_mapping)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

x = data2 [['Hours_Studied','Attendance','Sleep_Hours','Previous_Scores']] #Features
y = data2 [['Parental_Involvement_encoded']] #Target

#Split Data
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)

#Training
model = LogisticRegression()
model.fit(x_train, y_train)

#Evaluasi
y_pred = model.predict(x_test)

akurasi = accuracy_score(y_pred,y_test)
print(f'Hasil akurasi model = {akurasi:.2%}')

#apakah feature encoded bisa digunakan sebagai features? BISA
```

```python
import numpy as np
feature_importance = pd.DataFrame({
    'Feature': x.columns,
    'Importance': np.abs(model.coef_[0]) #Khusus klasifikasi, Pakai Absolute
})
feature_importance = feature_importance.sort_values(by='Importance',ascending = False)
print(feature_importance)
```
