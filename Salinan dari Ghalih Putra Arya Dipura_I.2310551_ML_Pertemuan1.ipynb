#pandas

digunakan untuk manipulasi data

```python
import pandas as pd
data = {"nama" : ['Andi', "Mulyono", "budi"] ,
        "umur" : [21, 30 , 35] ,
        "kota" :[ "Jakarta" , "Solo" , "Bogor"] }
data_2 = pd.DataFrame(data)
print(data_2)
```

```python
display(data_2)
```

```python
display(data_2[data_2.umur >= 30])
```

```python
display(data_2[data_2.nama == "Mulyono"])
```

```python
display(data_2[(data_2.nama == "Mulyono") | (data_2.nama == "budi")])
```
